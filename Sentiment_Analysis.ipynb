Sentiment Analysis : Contextual mining of text

*Sentiment Analysis is the most common text classification tool that analyses an incoming message and tells whether the underlying sentiment is positive, negative our neutral.*

**Document:** *By document, it is meant tweets, phrases, parts of news articles, whole news articles, a full article, a product manual, a story, etc.*

**Bag of Words**  *A bag-of-words representation of a document does not only contain specific words but all the unique words in a document and their frequencies of occurrences. The bag **does not contain any duplicate** words.*

Training data will look like : Capture.jpg

*The above representation is also known as* **Corpus.**

####  Naive Bayes Classification for Sentiment Analysis

*It is a classification technique based on Bayesâ€™ Theorem with an assumption of independence among predictors. In simple terms, a Naive Bayes classifier assumes that the presence of a particular feature in a class is unrelated to the presence of any other feature.*

**Bag of words assumption**: Assume position does not matter.

```python
# Load and prepare the dataset
import nltk
from nltk.corpus import movie_reviews
import random
```

```python
documents = [(list(movie_reviews.words(fileid)), category)
              for category in movie_reviews.categories()
              for fileid in movie_reviews.fileids(category)]

random.shuffle(documents)
```
